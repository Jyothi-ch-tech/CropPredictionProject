```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Crop_recommendation.csv")
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data=df.iloc[:,:7]
target=df.iloc[:,7:].values
```

```python
data
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
dataf=StandardScaler().fit_transform(data)
```

```python
X_train,X_test,y_train,y_test=train_test_split(dataf,target,test_size=0.2,random_state=42)
```

```python
dtc=DecisionTreeClassifier()
```

```python
model=dtc.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
new_soil_data = pd.DataFrame([[90, 42, 43, 20.879744, 82.002744, 6.502985, 202.935536]],
                             columns=['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall'])

prediction = model.predict(new_soil_data)
print("\n Predicted Crop for Given Soil Values:", prediction[0])
```

```python
print("\nDecision tree Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

# Train Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
print("\n Random Forest Accuracy:", accuracy_score(y_test, rf_pred))

# Train K-Nearest Neighbors
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
knn_pred = knn_model.predict(X_test)
print(" KNN Accuracy:", accuracy_score(y_test, knn_pred))

# Train Logistic Regression
lr_model = LogisticRegression(max_iter=200)  # Add max_iter for convergence
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
print(" Logistic Regression Accuracy:", accuracy_score(y_test, lr_pred))
```
